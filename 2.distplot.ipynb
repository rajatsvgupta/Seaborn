```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cars=sns.load_dataset("mpg").dropna()
```

```python
cars.head()
```

```python
sns.set_style("darkgrid")
```

```python
sns.distplot(cars.horsepower)
```

```python
sns.distplot(cars.horsepower,kde=False)
```

```python
sns.distplot(cars.horsepower,rug=True)
```

```python
sns.distplot(cars.horsepower,bins=15)
```

```python
sns.distplot(cars.weight,bins=[1500,2500,3500,5200])
```

```python
from scipy.stats import norm,skewnorm
```

```python
sns.distplot(cars.weight,fit=norm)
```

```python
sns.distplot(cars.weight,fit=skewnorm)
```

```python
sns.distplot(cars.weight,kde=False,hist_kws={"alpha":1})
```

```python
sns.distplot(cars.weight,kde=False,hist_kws={"alpha":1,"histtype":"step"})
```

```python
sns.distplot(cars.weight,kde=False,hist_kws={"alpha":1,"histtype":"stepfilled"})
```

```python
sns.distplot(cars.weight,hist=False)
```

```python
sns.distplot(cars.weight,kde_kws={"lw":5})
```

```python
sns.kdeplot(cars.weight,lw=5)
```

```python
sns.distplot(cars.weight,kde_kws={"bw":100})
```

```python
sns.distplot(cars.weight,kde_kws={"bw":1000})
```
