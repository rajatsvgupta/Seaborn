```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cars=sns.load_dataset("mpg").dropna()
```

```python
cars=cars[cars.cylinders.isin([4,6,8])]
```

```python
cars.head()
cars.dtypes
```

```python
sns.stripplot(cars.weight,cars.origin)
```

```python
sns.stripplot(cars.weight,cars.origin,hue=cars.cylinders)
```

```python
sns.stripplot(cars.weight,cars.origin,hue=cars.cylinders,dodge=True)
```

```python
sns.stripplot(cars.weight,cars.origin,jitter=False)
```

```python
sns.stripplot(cars.weight,cars.origin,jitter=0.1)
```

```python
from ipywidgets import interactive
```

```python
def jitter_widget(jitter=0.1):
    sns.stripplot(cars.weight,cars.origin,jitter=jitter)
```

```python
interactive(jitter_widget,jitter=(0,0.65,0.05))

```
