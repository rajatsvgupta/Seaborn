```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
diamonds=sns.load_dataset("diamonds")
```

```python
diamonds.head()
```

```python
diamonds=diamonds.sample(200,random_state=44)
```

```python
diamonds.shape
```

```python
sns.regplot(diamonds.carat,diamonds.price)
```

```python
sns.regplot(diamonds.carat,diamonds.price,fit_reg=False)
```

```python
sns.regplot(diamonds.carat,diamonds.price,scatter=False)
```

```python

plt.gca().set(xlim=(0,2.6))
sns.regplot(x='carat', y='price', data=diamonds, scatter=False);
```

```python
sns.regplot(x="carat",y="price",data=diamonds,ci=None)
```

```python
diamonds.head()
```

```python
cut_map={"Fair":1,
        "Good":2,
        "Very Good":3,
        "Premium":4,
        "Ideal":5}
diamonds["cut_value"]=diamonds.cut.map(cut_map)
```

```python
diamonds["cut_value"]
```

```python
diamonds.cut_value.value_counts()
```

```python
sns.regplot(x="cut_value",y="price",data=diamonds)
```

```python
sns.regplot(x="cut_value",y="price",data=diamonds,x_jitter=0.1)
```

```python
sns.regplot(x="cut_value",y="price",data=diamonds,x_estimator=np.mean)
```

```python
sns.regplot(x="cut_value",y="price",data=diamonds,x_estimator=np.mean,x_ci=50)
```

```python
sns.regplot(x="cut_value",y="price",data=diamonds,x_estimator=np.mean,x_ci=50)
```

```python
sns.regplot(x="carat",y="price",data=diamonds,fit_reg=False)
```

```python
sns.regplot(x="carat",y="price",data=diamonds)
```

```python
sns.regplot(x="carat",y="price",data=diamonds,order=2)
```
