```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
diamonds=sns.load_dataset("diamonds")
```

```python
diamonds.head()
```

```python
sns.set_style("dark")
```

```python
diamonds=diamonds[diamonds.cut.isin(["Premium","Good"])&
                 diamonds.color.isin(["D","F","J"])].sample(n=100,random_state=22)
```

```python
diamonds.head()
```

```python
sns.scatterplot(diamonds.carat,diamonds.price)
```

```python
sns.scatterplot("x","y",data=diamonds)
```

```python
sns.scatterplot("x","y",data=diamonds,hue="cut")
```

```python
sns.scatterplot("x","y",data=diamonds,hue="cut",palette=["purple","#55CCCC"])
```

```python
sns.scatterplot(x="carat",y="price",hue="depth",data=diamonds)
```

```python
sns.scatterplot(x="carat",y="price",style="cut",s=100,data=diamonds)
```

```python
sns.scatterplot(x="carat",y="price",size="cut",data=diamonds)
```

```python
sns.scatterplot(x="carat",y="price",size="cut",sizes=[150,80],data=diamonds)
```

```python
sns.scatterplot(x="carat",y="price",size="depth",data=diamonds)
```

```python
sns.scatterplot(x="carat",y="price",hue="cut",style="cut",data=diamonds,s=100)
```

```python
sns.scatterplot(x="carat",y="price",hue="cut",style="cut",size="depth",data=diamonds,s=100,sizes=(70,200))
```

```python
sns.scatterplot(x="carat",y="price",hue="cut",style="color",data=diamonds,s=100)
```

```python
sns.scatterplot(x="carat",y="price",data=diamonds,s=100,alpha=0.6)
```

```python
sns.scatterplot(x="carat",y="price",data=diamonds,s=200,marker="*")
```

```python
sns.scatterplot(x="carat",y="price",data=diamonds,s=200,marker="*",edgecolor="black")
```

```python
sns.scatterplot(x="carat",y="price",data=diamonds,s=200,marker="*",edgecolor="black",linewidth=2)
```
