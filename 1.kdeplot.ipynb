```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cars=sns.load_dataset('mpg').dropna()
```

```python
cars.head()
```

```python
cars.shape
```

Univariate Kdeplot

```python
sns.kdeplot(cars.horsepower)
```

```python
sns.kdeplot(cars.horsepower,shade=True)
```

```python
sns.kdeplot(cars.horsepower,bw=.1)
```

```python
sns.kdeplot(cars.horsepower,bw=4)
```

```python
sns.kdeplot(cars.horsepower,cumulative=True)
```

```python
from ipywidgets import interact, interactive

from IPython.display import display
```

```python
def bandwidth_widget(bw=10):
    sns.kdeplot(cars.horsepower, bw=bw, lw=3, shade=True)
    plt.xlim(-30, 300)
    plt.ylim(0, 0.03)
interactive(bandwidth_widget, bw=(1, 30, 1))
```

Bivariate Kde Plot

```python
sns.kdeplot(cars.horsepower, cars.mpg)
```

```python
sns.kdeplot(cars.horsepower, cars.mpg,n_levels=25)
```

```python
sns.kdeplot(cars.horsepower, cars.mpg, shade=True)
```

```python
sns.kdeplot(cars.horsepower, cars.mpg,shade=True,shade_lowest=False)
```

```python
sns.set_style('whitegrid')

sns.kdeplot(cars.horsepower, cars.mpg,
            shade=True,cbar=True,shade_lowest=False
           );
```
