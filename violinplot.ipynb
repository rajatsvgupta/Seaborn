```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cars=sns.load_dataset("mpg").dropna()
```

```python
cars.head()
```

```python
cars.cylinders.value_counts()
```

```python
cars=cars[cars.cylinders.isin([4,6,8])]
```

```python
cars.cylinders.value_counts()
```

```python
sns.violinplot(x=cars.cylinders,y=cars.displacement)
```

```python


sns.violinplot(x=cars.cylinders,y=cars.displacement,hue=cars.origin)
```

```python
x=cars[cars.origin.isin(["japan","europe"])]
x
```

```python
sns.violinplot(x="cylinders",y="displacement",hue="origin",data=x)
plt.legend(loc=2)
```

```python
sns.violinplot(x="cylinders",y="displacement",hue="origin",data=x,split=True)
plt.legend(loc=2)
```

```python
sns.violinplot(x="cylinders",y="displacement",hue="origin",data=x,split=True,inner="quartiles")
plt.legend(loc=2)
```

```python
sns.violinplot(x="cylinders",y="displacement",hue="origin",data=x,split=True,scale="count")
plt.legend(loc=2)
```

```python
sns.violinplot(x="cylinders",y="displacement",hue="origin",data=x,split=True,scale="count",scale_hue=False)
plt.legend(loc=2)
```

```python
sns.violinplot(x="origin",y="displacement",data=cars,order=["japan","europe","usa"])
```

```python
sns.violinplot(x="origin",y="displacement",data=cars,order=["japan","europe","usa"],linewidth=3)
```

```python
sns.violinplot(x="origin",y="displacement",data=cars,order=["japan","europe","usa"],bw=3)
```

```python
sns.violinplot(x="origin",y="displacement",data=cars,order=["japan","europe","usa"],bw=.3)
```

```python
sns.violinplot(y="displacement",data=cars,cut=3)
```
