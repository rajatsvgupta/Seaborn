```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
 cars=sns.load_dataset("mpg")
```

```python
cars.head()
```

```python
cars.groupby("origin").cylinders.value_counts()
```

```python
orig_cy=cars.groupby("origin").cylinders.value_counts().unstack()
```

```python
sns.heatmap(orig_cy)
```

```python
orig_cy=cars.groupby("origin").cylinders.value_counts().unstack().fillna(0)
```
